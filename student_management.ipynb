```python
class Student:
    def __init__(self,stu_id,name,age):
        self.stu_id = stu_id
        self.name = name
        self.age = age
    
class Student_List:
    def __init__(self):
        self.students = []
    def main(self):
        print('======= STUDENT MANAGEMENT PROGRAM =======')
        print('1. Student Information\n2. Add student\n3. Update student\n4. Delete student\n5. Lookup student\n6. Exit')
        while True:
            user_choice = input('Select function: ')            
            if user_choice == '1':
                print('======= STUDENT INFORMATION =======')
                self.info(self.students)
            elif user_choice == '2':
                print('======= ADD STUDENT =======')
                self.add()
            elif user_choice == '3':
                print('======= UPDATE STUDENT =======')
                self.update()
            elif user_choice == '4':
                print('======= DELETE STUDENT =======')
                self.delete()
            elif user_choice == '5':
                print('======= LOOKUP STUDENT =======')
                self.lookup()
            elif user_choice == '6':
                print('======= GOODBYR =======')
                break
                
    def info(self,students):
        for student in self.students:
            print(f'Student {student.stu_id}, name: {student.name}, age: {student.age}')           
            
    def add(self):
        run = True
        while run:
            new_stu_id = input('Input student\'s id: ')
            if self.students == []:
                run = False                
            else:
                for student in self.students:
                    if student.stu_id == new_stu_id:
                        print('Student\'s id is existed!')
                    else:
                        run = False     
        name = input('Input student\'s name: ')
        age = input('Input student\'s age: ')
        student = Student(new_stu_id,name,age)
        self.students.append(student)
        print('Add successful!')
        pass
    
    def update(self):
        update_id = input('Input student\'s id to update: ')
        for student in self.students:
            if student.stu_id == update_id:
                new_name = input('Update student\'s name: ')
                new_age = input('Update student\'s age: ')
                student.name = new_name
                student.age = new_age
                print('Update successful!') 
            else:
                print('Student not founded!')
    
    def delete(self):
        delete_id = input('Input student\'s id to delete: ')
        for student in self.students:
            if student.stu_id == delete_id:
                self.students.pop(self.students.index(student))
                print('Delete successful!') 
            else:
                print('Student not founded!')
                
    def lookup(self):
        lookup_id = input('Input student\'s id to lookup:')
        for student in self.students:
            if student.stu_id == lookup_id:
                print(f'Student {student.stu_id}, name: {student.name}, age: {student.age}') 
                return
        else:
            print('Student not founded!')

student = Student_List()
student.main()
```
